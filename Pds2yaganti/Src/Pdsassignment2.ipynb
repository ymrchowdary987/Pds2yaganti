```python
import pandas as pd # importing pandas library
```

```python
datacar=pd.read_csv("/content/train.csv") # reading the csv file located in datacar
```

```python
datacar.head()
```

```python
datacar.isna().sum() #checked for null values
```

```python
datacar2 = pd.DataFrame(datacar)
```

```python
# b. Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).
```

```python
datacar2['Mileage'] = datacar2['Mileage'].str.extract('(\d+\.\d+)').astype(float)

datacar2['Engine'] = datacar2['Engine'].str.replace(' CC', '').astype(float)

datacar2['Power'] = datacar2['Power'].str.extract('(\d+\.\d+)').astype(float)

datacar2['New_Price'] = datacar2['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
# a. Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task.
```

```python
datacar2['Mileage'].fillna(datacar2['Mileage'].median(),inplace=True)
datacar2['Engine'].fillna(datacar2['Engine'].median(),inplace=True)
datacar2['Power'].fillna(datacar2['Power'].median(),inplace=True)

datacar2['Seats'].fillna(datacar2['Seats'].mode()[0],inplace=True)

datacar2.drop('New_Price', axis=1, inplace=True)
```

```python
outputcsv_file = 'cleandata.csv'
datacar2.to_csv(outputcsv_file, index=False)
```
